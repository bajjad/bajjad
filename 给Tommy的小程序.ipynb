```python
#-*- coding: utf-8 -*-
#   格式：\033[显示方式;前景色;背景色m
#   说明:
#
#   前景色            背景色            颜色
#   ---------------------------------------
#     30                40              黑色
#     31                41              红色
#     32                42              绿色
#     33                43              黃色
#     34                44              蓝色
#     35                45              紫红色
#     36                46              青蓝色
#     37                47              白色
#
#   显示方式           意义
#   -------------------------
#      0           终端默认设置
#      1             高亮显示
#      4            使用下划线
#      5              闪烁
#      7             反白显示
#      8              不可见
#
#   例子：
#   \033[1;31;40m    <!--1-高亮显示 31-前景色红色  40-背景色黑色-->
#   \033[0m          <!--采用终端默认设置，即取消颜色设置-->]]]
 
STYLE = {
        'fore':
        {   # 前景色
            'black'    : 30,   #  黑色
            'red'      : 31,   #  红色
            'green'    : 32,   #  绿色
            'yellow'   : 33,   #  黄色
            'blue'     : 34,   #  蓝色
            'purple'   : 35,   #  紫红色
            'cyan'     : 36,   #  青蓝色
            'white'    : 37,   #  白色
        },
 
        'back' :
        {   # 背景
            'black'     : 40,  #  黑色
            'red'       : 41,  #  红色
            'green'     : 42,  #  绿色
            'yellow'    : 43,  #  黄色
            'blue'      : 44,  #  蓝色
            'purple'    : 45,  #  紫红色
            'cyan'      : 46,  #  青蓝色
            'white'     : 47,  #  白色
        },
 
        'mode' :
        {   # 显示模式
            'mormal'    : 0,   #  终端默认设置
            'bold'      : 1,   #  高亮显示
            'underline' : 4,   #  使用下划线
            'blink'     : 5,   #  闪烁
            'invert'    : 7,   #  反白显示
            'hide'      : 8,   #  不可见
        },
 
        'default' :
        {
            'end' : 0,
        },
}

def style(string, mode = '', fore = '', back = ''): 
    mode = '%s' % STYLE['mode'][mode] if mode in STYLE['mode'] else ''
    fore = '%s' % STYLE['fore'][fore] if fore in STYLE['fore'] else ''
    back = '%s' % STYLE['back'][back] if back in STYLE['back'] else ''
    style = ';'.join([s for s in [mode, fore, back] if s])
    style = '\033[%sm' % style if style else ''
    end = '\033[%sm' % STYLE['default']['end'] if style else ''
    return '%s%s%s' % (style, string, end)
```

```python
## 1# 给Tommy的小程序 复杂版
# 根据输入生成 数字1 ✖️或者➗ 数字2 +或者- 数字3
# 例子：
#    36 / 4 + 27 =
#    6 * 6 - 3 =

# 导入random中的两个函数，randint产生整数, choice随机抽取列表中任意一个元素
from random import randint, choice, shuffle
from datetime import datetime
import re
import sys

def generate(num=1):
    list1 = []
    
    # 根据输入控制循环次数
    for i in range(1,num+1):

        # 乘数1
        mult1 = randint(2, 20)
        # 乘数2
        mult2 = randint(2, 10)

        # 第三个数
        mult3 = randint(1, 110)

        # 算出乘数的积
        div1 = mult1 * mult2

        # 先取乘除
        opt1 = ['*','/']

        option = choice(opt1)

        # 如果是除法，乘数1改成积
        if option == '/':
            mult1 = div1

        # 生成公式1
        s = str(mult1) + ' ' + option + ' ' + str(mult2)
        
        
        # 计算公式1的得数
        result = eval(s.replace(' ', '')) # 字符串的replace函数请单独练习

        # 再取加减
        opt2 = ['+', '-']

        option = choice(opt2)

        # 如果是减法，把大数放在前面
        if re.search('/|-', option) and mult3 > result:
            s, mult3 = mult3, s

        # 生成公式2
        s = str(s) + ' ' + option + ' ' + str(mult3)

        list1.append(s)
        
    return list1
    
# 红字输出函数，只是美观
def red(word="Hello"):
    return '\033[1;31m{}\033[0m'.format(word)

# 程序健壮性处理
while True:
    name = input("请输入你的名字：")
    if name.isalpha():
        break
    else:
        print("输入错误。")
        

print( style(name.title(), mode = 'bold',fore='blue') +"小朋友，欢迎你进入" + style('奇妙', fore='blue') + "的" + style('数', fore='red') + style('学', fore='yellow') + style('世', fore='blue') + style('界',fore='purple') + "。")

errors = []

# 程序健壮性处理
while True:
    try:
        num = int(input("请问你要做几道题呀："))
        break
    except:
        print("输入错误。")

# 分数
score = 0

start = datetime.now()

list1 = generate(num)

for i in range(1, num+1):
    
    s = list1.pop()
    result = eval(s.replace(' ', ''))
    
    # 程序健壮性处理
    while True:
        try:
            answer=float(input('第'+str(i) + "题. 请问："+s.replace('*','✖️').replace('/','➗')+" = "))
            break
        except:
            print("输入错误。")

    # 判断答案是否正确
    if answer == result:
        score += 1
        print(style("你答对了！", fore='blue'))
    else:
        temp = list(str(int(result)))
        temp[randint(0,len(temp)-1)] = '*'
        temp = ''.join(temp)
        
        print(style("你答错了。", fore='red') + '正确答案是：{}'.format(style(temp,fore='blue')))
        errors.append(s)
        
        
# 计算正确率
if num > 0:
    lv = score / num
else:
    lv = 0

print()
print("感谢你使用本程序，你答对了{}道题。".format(style(score,fore='red')))

if lv > 0.8:
    s = '达到'
    answer = '太棒了'
else:
    s = '只有'
    answer = '还要努力哦'
    
r = str(int(lv*100)) + '%'
print("正确率{}".format(s) + style(r,fore='red') + '，{}！'.format(answer))

end = datetime.now()

time_used = round(float((end - start).seconds)/60, 2)

print('你用了{}'.format(time_used) + '分钟。')

if errors:
    print()
    answer = input('共有{}道错题，是否现在补做错的题(yes/no): '.format(len(errors)))
    if answer.lower() in 'yes':
        b = [i for i in range(0,10)]
        shuffle(b)
        
        while True:
            a = input('上天给了你一次多做几道题的机会，有彩蛋哦！请输入1-10进入盲选：')
            try:
                if 0 < int(a) < 11:
                    break
            except:
                pass
        
        if b[int(a)-1] == 0:
            print('恭喜你，你是'+style("天选之子！", fore='red')+'你抽中了超级大礼包，不用做题了，直接下课，回家。Goodluck！')
            sys.exit(0)
        else:

            print('恭喜你，获得了多做{}道题的机会，加油！我看好你哦！'.format(style(str(b[int(a)-1]),fore='red')))
            for i in range(1, b[int(a)-1]+1):
                print('{0}道题啊{0}道题。'.format(style(str(b[int(a)-1]),fore='blue')))
                print()
            add = generate(b[int(a)-1])

            errors.extend(add)
            print()

            i = 1

            while errors:
                shuffle(errors)
                s = errors.pop()
                result = eval(s.replace(' ', ''))

                # 程序健壮性处理
                while True:
                    try:
                        answer=float(input('第'+str(i) + "题. 请问："+s+" = "))
                        break

                    except:
                        print("输入错误。")

                # 判断答案是否正确
                if answer == result:
                    print(style("你答对了！", fore='blue'))
                else:
                    print(style("你答错了。", fore='red'))
                    errors.append(s)
                i += 1
```
